## Init model

```python
from ultralytics import YOLOv10

MODEL_PATH = "yolov10n.pt"
model = YOLOv10(MODEL_PATH)
```

```python
model.info()
```

```python
import cv2
import matplotlib.pyplot as plt

IMG_PATH = "./images/street.png"
img = cv2.imread(IMG_PATH)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
result = model(source=IMG_PATH)[0]
result
```

```python
result.save("./images/street_predicted.png")
```

```python
img = cv2.imread("./images/street_predicted.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```
